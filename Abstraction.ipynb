```python
from abc import ABC, abstractmethod
```

```python
class Employee(ABC):
    @abstractmethod
    def work(self):
        pass

class Developer(Employee):
    def work(self):
        print("Writing Code")

class Tester(Employee):
    def work(self):
        print("Testing code")

class Manager(Employee):
    def work(self):
        print("Manages employee")

dev = Developer()
test = Tester()
manager = Manager()


employees = [dev, test, manager]

for emp in employees:
    emp.work()
```
